```python
# Read the data from the CSV file into a DataFrame and display the data
pd.set_option('display.max_columns', None) # Display alone does not work because there are too many columns so force display all columns
data = pd.read_csv('new_train_EGN5442.csv')
originalData = data # Storing the unmodified data for later use
display(data)


'''
Initial observations:
- The unnamed column looks like a copy of the index so not needed
- x2 is # followed by a bunch of numbers
- x5 - x23 nothing stands out yet
- x24 floats instead of ints
- x25 and x26 look like binary values (D_C or L_C) and (Bt or PT)
- x29 string with opening "(", numbers, and then "&&)" to end the string
- x30 looks like a column with no useful information
- x31 has a lot of rows with value "inf" so will have to deal with it later on
- x4 the entire column is 4
- x6 nothing out of the ordinary
'''
```

```python
'''
Before working on cleaning the data, I am going to remove
all of the columns that do not have any useful data for logistic regression.

x30 is NaN for the entire column so it will not be useful for regression
x4 is all 4 so there is no need to use it for regression
The unnamed: 0 column repeates the index in the table which is also unnecessary for regression
'''
display(data[["Unnamed: 0", "x4", "x30"]])
```

```python
# Removing the three unnecessary columns
data = data.drop("Unnamed: 0", axis = 1)
data = data.drop("x4", axis = 1)
data = data.drop("x30", axis = 1)
display(data)
```

```python
'''
Are there any string or datetime columns? How to deal with them? Use the apply() with
user-defined functions (or lambda expressions) when appropriate.

Based on what was found above the only columns that need cleaned are "x2", "x3", "x14", "x25", "x26", "x29"
'''

def cleanX2(value):
  # Each string in x2 is a # followed by a number so just remove the # and convert string to int
  return int(value[1:])
data["x2"] = data["x2"].apply(cleanX2) # Apply cleanX2 to each row

def cleanX3(gender):
  # This is just a binary column so set Male to 1 and Female to 0
  if gender == "Male":
    return 1
  else:
    return 0
data["x3"] = data["x3"].apply(cleanX3) # Apply cleanX3 to each row

def cleanX14(value): # Convert each of the strings to int
  # invalid literal for int() with base 10: ' ' so theres some empty strings that need fixed
  if value == " ":
    return np.nan # Clean these rows later on
  else:
    return int(value)
data["x14"] = data["x14"].apply(cleanX14) # Apply cleanX14 to each row

def cleanX25(value): # Convert L_C and D_C to 0 and 1
  if (value == "L_C"):
    return 1
  else:
    return 0
data["x25"] = data["x25"].apply(cleanX25) # Apply cleanX25 to each row

def cleanX26(value): # Convert Bt and PT to 0 and 1
  if (value == "PT"):
    return 1
  else:
    return 0
data["x26"] = data["x26"].apply(cleanX26) # Apply cleanX26 to each row

def cleanX29(value): # Look at the string and add only the ints to a new string and return
  intString = ""
  for char in value:
    if char.isdigit():
      intString += char
  return int(intString)
data["x29"] = data["x29"].apply(cleanX29) # Apply cleanX29 to each row

display(data[["x2", "x3", "x14", "x25", "x26", "x29"]])
```

```python
data.isna() # Check for missing data
```

```python
data.isna().sum() # Find the total missing data per column
# Looks like x14 had a lot of empty strings and x24 is also missing some values
```

```python
# https://stats.stackexchange.com/questions/143700/which-is-better-replacement-by-mean-and-replacement-by-median
# For x24 and x14 since the columns do not contain extreme outliers I will replace missing values with mean
# There are alot of missing values in x14 so I will revisit if this has a negative impact on the models
data["x14"] = data["x14"].fillna(data["x14"].mean())
data["x24"] = data["x24"].fillna(data["x24"].mean())
display(data)
```

```python
data.isna().sum()
```

```python
# Now that there are no missing values the last thing to do is to deal with the inf values in x31
def cleanX31(value): # Convert inf to nan then nan to mean
  if (value == np.inf):
    return np.nan
  else:
    return value
data["x31"] = data["x31"].apply(cleanX31) # Apply cleanX31 to each row
data["x31"] = data["x31"].fillna(data["x31"].mean())
```
